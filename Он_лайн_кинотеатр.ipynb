# Python


```
import pandas as pd
```

```
colnames_r = ['user_id', 'item_id', 'rating', 'timestamp']
```

```
df_ratings  = pd.read_csv('u.data.csv', names=colnames_r, header=None, sep='\t', encoding = 'ISO-8859-1')
```

```
df_ratings.head()
```

```
colnames_m = ['movie_id', 'movie_title', 'release_date', 'video_release_date', 
              'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 
              'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
              'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller',
              'War', 'Western']
```

```
df_movies  = pd.read_csv('u.item.csv', names=colnames_m, header=None, sep='|', encoding = 'ISO-8859-1')
```

```
df_movies.head()
```

```
df_ratings[['user_id', 'rating']].groupby(by='user_id').count().sort_values('rating', ascending=False).head(1)
```

```
df_ratings_top = df_ratings[df_ratings.user_id == 405]
```

```
df_ratings_top.head()
```

```
df_ratings_top.info()
```

```
df_ratings_genres = df_ratings_top.merge(df_movies, how = 'inner', left_on = 'item_id', right_on = 'movie_id')
```

```
df_ratings_genres.head()
```

```
df_count = df_ratings[['item_id', 'rating']].groupby(by='item_id').count().sort_values('rating', ascending=False)
```

```
df_count.head()
```

```
df_sum = df_ratings[['item_id', 'rating']].groupby(by='item_id').sum().sort_values('rating', ascending=False)
```

```
df_sum.head()
```

```
df_count_sum = df_count.merge(df_sum, how = 'inner', left_on = 'item_id', right_on = 'item_id')
```

```
df_count_sum.head()
```

```
df_ratings_genres_count_sum = df_ratings_genres.merge(df_count_sum, how = 'inner', left_on = 'item_id', right_on = 'item_id')
```

```
df_ratings_genres_count_sum.head()
```

```
df_ratings_genres_count_sum.rename(columns={'rating_x': 'count', 'rating_y': 'sum'}, inplace=True)
```

```
df_ratings_genres_count_sum['year'] = pd.DatetimeIndex(df_ratings_genres_count_sum['release_date']).year
```

```
df_ratings_genres_count_sum.head()
```

```
df_ratings_genres_count_sum.columns
```

```
df_modeling = df_ratings_genres_count_sum[['unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'count', 'sum',
       'year']]
```

```
df_modeling.head()
```

```
X, y = df_modeling[['unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'count', 'year']], df_modeling['sum']
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
sc = StandardScaler()
```

```
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
model = LinearRegression()
```

```
model.fit(X_train, y_train)
```

```
mean_squared_error(y_train, model.predict(X_train))
```

```
mean_squared_error(y_test, model.predict(X_test))
```

```
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```
df_ratings_spark = spark.read.csv('u.data.csv', inferSchema=True, sep='\t', 
                        encoding='ISO-8859-1').toDF('user_id', 'item_id', 'rating', 'timestamp')
```

```
df_ratings_spark.show()
```

```
df_movies_spark = spark.read.csv('u.item.csv', inferSchema=True, sep='|', 
                        encoding='ISO-8859-1').toDF('movie_id', 'movie_title', 'release_date', 'video_release_date', 
              'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 
              'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film_Noir',
              'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi', 'Thriller',
              'War', 'Western')
```

```
df_movies_spark.show()
```

```
df_avg = df_ratings_spark.orderBy('item_id').groupby('item_id').avg('rating')
df_avg.show()
```

```
total = df_ratings_spark.join(df_movies_spark, df_ratings_spark['item_id']==df_movies_spark['movie_id'], how='inner')
```

```
total.show()
```

```
df_genres = total.select('unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 
              'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film_Noir',
              'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi', 'Thriller',
              'War', 'Western', 'rating')
df_genres.show()
```

```
df_unknown = df_genres.filter("unknown == '1'").select('unknown', 'rating').groupBy('unknown').avg('rating').show()
df_action = df_genres.filter("Action == '1'").select('Action', 'rating').groupBy('Action').avg('rating').show()
df_adventure = df_genres.filter("Adventure == '1'").select('Adventure', 'rating').groupBy('Adventure').avg('rating').show()
df_animation = df_genres.filter("Animation == '1'").select('Animation', 'rating').groupBy('Animation').avg('rating').show()
df_childrens = df_genres.filter("Childrens == '1'").select('Childrens', 'rating').groupBy('Childrens').avg('rating').show()
df_comedy = df_genres.filter("Comedy == '1'").select('Comedy', 'rating').groupBy('Comedy').avg('rating').show()
df_crime = df_genres.filter("Crime == '1'").select('Crime', 'rating').groupBy('Crime').avg('rating').show()
df_documentary = df_genres.filter("Documentary == '1'").select('Documentary', 'rating').groupBy('Documentary').avg('rating').show()
df_drama = df_genres.filter("Drama == '1'").select('Drama', 'rating').groupBy('Drama').avg('rating').show()
df_fantasy = df_genres.filter("Fantasy == '1'").select('Fantasy', 'rating').groupBy('Fantasy').avg('rating').show()
df_film_noir = df_genres.filter("Film_Noir == '1'").select('Film_Noir', 'rating').groupBy('Film_Noir').avg('rating').show()
df_horror = df_genres.filter("Horror == '1'").select('Horror', 'rating').groupBy('Horror').avg('rating').show()
df_musical = df_genres.filter("Musical == '1'").select('Musical', 'rating').groupBy('Musical').avg('rating').show()
df_mystery = df_genres.filter("Mystery == '1'").select('Mystery', 'rating').groupBy('Mystery').avg('rating').show()
df_romance = df_genres.filter("Romance == '1'").select('Romance', 'rating').groupBy('Romance').avg('rating').show()
df_sci_fi = df_genres.filter("Sci_Fi == '1'").select('Sci_Fi', 'rating').groupBy('Sci_Fi').avg('rating').show()
df_thriller = df_genres.filter("Thriller == '1'").select('Thriller', 'rating').groupBy('Thriller').avg('rating').show()
df_war = df_genres.filter("War == '1'").select('War', 'rating').groupBy('War').avg('rating').show()
df_western = df_genres.filter("Western == '1'").select('Western', 'rating').groupBy('Western').avg('rating').show()
```
